### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



# Ingresos totales por cada tienda ✨

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
import matplotlib.pyplot as plt
#Ingresos totales por tienda
ingT1 = tienda['Precio'].sum()
ingT2 = tienda2['Precio'].sum()
ingT3 = tienda3['Precio'].sum()
ingT4 = tienda4['Precio'].sum()
print(f'El ingreso total de la Tienda 1 es: {ingT1}')
print(f'El ingreso total de la Tienda 2 es: {ingT2}')
print(f'El ingreso total de la Tienda 3 es: {ingT3}')
print(f'El ingreso total de la Tienda 4 es: {ingT4}')
```

```python
#Visualización de datos en Gráficas

#----Gráfica lineal----
#Datos
x = ['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
y = [ingT1,ingT2,ingT3,ingT4]

#Creación del gráfico lineal
plt.plot(x,y)

#Añadir etiquetas
plt.title("Gráfico de ingresos totales")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

#Mostrar
plt.show()

#----Gráfica de barras----

# Datos para el gráfico
tiendas = ['1', '2', '3', '4']
ingresos = [ingT1, ingT2, ingT3, ingT4]

# Crear el gráfico de barras
plt.bar(tiendas, ingresos)

# Añadir etiquetas y título
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.title('Gráfico de ingresos totales')
for i, v in enumerate(ingresos):
  plt.text(i, v, f'{v}', ha='center', va='bottom')

#Mostrar
plt.show()

#----Gráfica de pastel----

#Datos para el gráfico
tiendas = [1,2,3,4]
ingresos = [ingT1,ingT2,ingT3,ingT4]

#Añadir etiquetas y título
plt.title('Gráfico de ingresos totales')
plt.pie(ingresos, labels=tiendas)


#Mostrar
plt.show()
```

# 2. Ventas por categoría

#Ventas por categoría ✨
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
#Primeras tres categorías populares de la tienda 1
venCatT1 = tienda.groupby('Categoría del Producto')['Producto'].count()
print(f'Categorías más populares de la tienda 1: {venCatT1.head(3)}')

#Primeras tres categorías populares de la tienda 2
venCatT2 = tienda2.groupby('Categoría del Producto')['Producto'].count()
print(f'Categorías más populares de la tienda 2: {venCatT2.head(3)}')

#Primeras tres categorías populares de la tienda 3
venCatT3 = tienda3.groupby('Categoría del Producto')['Producto'].count()
print(f'Categorías más populares de la tienda 3: {venCatT3.head(3)}')

#Primeras tres categorías populares de la tienda 1
venCatT4 = tienda4.groupby('Categoría del Producto')['Producto'].count()
print(f'Categorías más populares de la tienda 1: {venCatT4.head(3)}')
```

# 3. Calificación promedio de la tienda


#Valoración media por tienda ✨
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.


```python
#Extracción de datos y cálculo del promedio
calPromT1 = tienda['Calificación'].mean()
calPromT2 = tienda2['Calificación'].mean()
calPromT3 = tienda3['Calificación'].mean()
calPromT4 = tienda4['Calificación'].mean()

#Mostrar promedio redondedado a 2 cifras
print(f'La calificación promedio de la tienda 1 es: {calPromT1:.2f}')
print(f'La calificación promedio de la tienda 2 es: {calPromT2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {calPromT3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {calPromT4:.2f}')
```

# 4. Productos más y menos vendidos

#Productos más y menos vendidos ✨
En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
#Función para obtener los productos
def prodVend(tienda):
  contProd = tienda['Producto'].value_counts()
  prodTop = contProd.head(3)
  prodLow = contProd.tail(3)
  return prodTop, prodLow

#Utilización de la función para cada tienda
prodTopT1, prodLowT1 = prodVend(tienda)
prodTopT2, prodLowT2 = prodVend(tienda2)
prodTopT3, prodLowT3 = prodVend(tienda3)
prodTopT4, prodLowT4 = prodVend(tienda4)

#Visualización de los resultados
print (f'Los 3 produtos más vendidos de la tienda 1 son: {prodTopT1}\nLos 3 produtos menos vendidos de la tienda 1 son: {prodLowT1}\n\n')
print (f'Los 3 produtos más vendidos de la tienda 2 son: {prodTopT2}\nLos 3 produtos menos vendidos de la tienda 2 son: {prodLowT2}\n\n')
print (f'Los 3 produtos más vendidos de la tienda 3 son: {prodTopT3}\nLos 3 produtos menos vendidos de la tienda 3 son: {prodLowT3}\n\n')
print (f'Los 3 produtos más vendidos de la tienda 4 son: {prodTopT4}\nLos 3 produtos menos vendidos de la tienda 4 son: {prodLowT4}\n\n')
```

# 5. Envío promedio por tienda

#Valor del envío promedio por tienda ✨
En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.



```python
#Función que calcula el costo de envío promedio
def costPromEnv(envio):
  return envio['Costo de envío'].mean()

#Utilización de la función por tienda
envT1 = costPromEnv(tienda)
envT2 = costPromEnv(tienda2)
envT3 = costPromEnv(tienda3)
envT4 = costPromEnv(tienda4)

#Mostrar los reesultados redondeado a dos números decimales
print(f'El costo de envío promedio de la tienda 1 es: {envT1:.2f}')
print(f'El costo de envío promedio de la tienda 2 es: {envT2:.2f}')
print(f'El costo de envío promedio de la tienda 3 es: {envT3:.2f}')
print(f'El costo de envío promedio de la tienda 4 es: {envT4:.2f}')
```
